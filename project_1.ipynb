```python
# 1.Load A Data
import numpy as np
import pandas as pd 
crop_yield = pd.read_csv(r"C:\Users\Acer\OneDrive\Desktop\yield_df.csv")
crop_yield
```

```python
crop_yield.isna().sum()
```

```python
# 2.Feature and Target Identification 
a = crop_yield [["Year", "average_rain_fall_mm_per_year", "pesticides_tonnes","avg_temp"]]
b= crop_yield [["hg/ha_yield"]]
print(a)
```

```python
# 3.Splitting Data 
from sklearn.model_selection import train_test_split
a_train , a_test , b_train , b_test  = train_test_split( a,b,test_size=0.25 , random_state=42 )
```

```python
# 4.Model Selection and Training 
from sklearn.linear_model import LinearRegression 
model = LinearRegression()
model.fit(a_train ,b_train)
```

```python
# 5.Prediction OR Testing 
b_pred = model.predict(a_test)
b_pred
```

```python
a_test
```

```python
# 6.Evaluation 
from sklearn.metrics import r2_score 
R2 = r2_score(b_test , b_pred )
print(R2* 100)
```

```python
# 7.Saving Model 

import joblib as jb 
jb.dump(model , "crop_yield_prediction_Output_Cal.pkl")
```
